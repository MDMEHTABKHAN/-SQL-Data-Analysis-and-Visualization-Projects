# Common Plot Types in Matplotlib

1. Line Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y, color = 'b', marker = 'o', linestyle='--')
plt.title('Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Bar Plot
Vertical bar char

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

plt.bar(categories, values, color= 'orange')
plt.title('vertical bar plot')
plt.show()
```

Horizontal bar chart

```python
plt.barh(categories, values, color='green')
plt.title("Horizontal Bar Plot")
plt.show()
```

3. Histogram

Used for frequency distribution.

```python
import numpy as np
```

```python
data = np.random.randn(1000)  
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram")
plt.show()
```

4. Scatter Plot

Used to show relationship between two variables.

```python
x = [5, 7, 8, 7, 6, 9, 5, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

5. Pie Chart

Used to show proportions.

```python
sizes = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['red', 'green', 'blue', 'orange', 'purple']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

6. Stack Plot

Used to show cumulative values over time.

```python
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating   = [2, 3, 4, 3, 2]
working  = [7, 8, 7, 2, 2]
playing  = [8, 5, 7, 8, 13]

plt.stackplot(days, sleeping, eating, working, playing, 
              labels=['Sleeping', 'Eating', 'Working', 'Playing'],
              colors=['m','c','r','g'])
plt.legend(loc='upper left')
plt.title("Stack Plot")
plt.show()
```

Area Plot

Basically a filled line plot.

```python
x = range(1, 6)
y = [1, 4, 6, 8, 10]

plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.plot(x, y, color="blue")
plt.title("Area Plot")
plt.show()
```
