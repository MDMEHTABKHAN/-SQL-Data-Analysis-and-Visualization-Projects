```python
import matplotlib.pyplot as plt
```

```python
x = (1,2,3,4)
y = (10,20,30,40)

plt.plot(x,y)
plt.show()
```

```python
plt.plot([10,20,30,40,50], [100,200,300,400,500]) # x,y
plt.show()
```

```python
x = (10,20,30)
y  = (100,200,300)
z = (1000,2000,3000)
plt.plot(x,y,z)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y, color='r', linestyle='--', marker='o')  #('r-o') r=red, --=dashed, o=marker
plt.title("Line plot with matplotlib")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 4, 9, 16]

plt.plot(x, y1, label="Linear")
plt.plot(x, y2, label="Quadratic")

plt.title("Example Plot")        
plt.xlabel("X-axis")             
plt.ylabel("Y-axis")             
plt.legend()        
plt.show()
```
