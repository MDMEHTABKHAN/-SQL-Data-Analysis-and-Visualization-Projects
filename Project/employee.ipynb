```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data
data = {
    "EmployeeID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jane"],
    "Department": ["HR", "IT", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 60000, 65000, 70000, 52000, 62000, 71000, 53000, 64000, 72000],
    "YearsAtCompany": [2, 5, 3, 7, 1, 4, 8, 2, 5, 9]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df.head(10))
print(df.tail(10))
# Basic statistics
print("\nBasic Statistics:")
print(df.describe())
df.info()
```

```python
# Average salary by department
avg_salary = df.groupby("Department")["Salary"].mean()
print("\nAverage Salary by Department:")
print(avg_salary)
```

```python
#  count employeees in each department
dept_counts = df['Department'].value_counts()
print("\nEmployee Count by Department:")
print(dept_counts)
```

```python
# maximum values salary by department
max_salary = df.groupby("Department")["Salary"].max()
print("\nMaximum Salary by Department:")
print(max_salary)
```

```python
# minimum salary by department
min_salary = df.groupby('Department')['Salary'].min()
print("\nMinimum Salary by Department:")
print(min_salary)
```

```python
# total salary by department
total_salary = df.groupby('Department')['Salary'].sum()
print("\nTotal Salary by Department:")
print(total_salary)
```

```python
highest_salary = df[df["Salary"] == df["Salary"].max()]
lowest_salary = df[df["Salary"] == df["Salary"].min()]

print("Highest Salary:\n", highest_salary)
print("Lowest Salary:\n", lowest_salary)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Salary"], bins=5, kde=True, color='skyblue')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Department", data=df, palette="Set2")
plt.title("Number of Employees per Department")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=avg_salary.index, y=avg_salary.values, palette="Set3")
plt.title("Average Salary by Department")
plt.ylabel("Average Salary")
plt.show()
```
