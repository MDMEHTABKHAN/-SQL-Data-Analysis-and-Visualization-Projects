```python
import pandas as pd
```

```python
data = {
    "Student": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Math": [85, 78, 92, 70, 88],
    "Science": [90, 80, 85, 75, 95],
    "English": [78, 85, 88, 82, 90]
}


df = pd.DataFrame(data)
print("Dataset:")
df
```

```python

print("\n Info:")
print(df.info())

print("\n Statistics:")
print(df.describe())


print("\n Math Scores:")
print(df["Math"])

print("\n Students with Math > 80:")
print(df[df["Math"] > 80])

df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)
print("\n DataFrame with Average column:")
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
# Bar chart: Average scores of students
plt.bar(df["Student"], df["Average"], color="black")
plt.title("Average Scores of Students")
plt.xlabel("Students")
plt.ylabel("Average Score")
plt.show()
```

```python
# Line plot: Subject-wise performance
plt.plot(df["Student"], df["Math"], marker="o", label="Math")
plt.plot(df["Student"], df["Science"], marker="s", label="Science")
plt.plot(df["Student"], df["English"], marker="^", label="English")

plt.title("Subject-wise Performance")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend()
plt.show()
```

```python
# Boxplot: Distribution of scores
plt.boxplot([df["Math"], df["Science"], df["English"]],
            labels=["Math", "Science", "English"])
plt.title("Score Distribution")
plt.show()
```
